```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
performance_data = [pd.read_csv(i) for i in os.listdir()]
```

```
plt.figure(figsize=(12, 6))
ax=plt.subplot(1,1,1,)
for i in range(len(performance_data)):
  ax.plot(performance_data[i].acc[15:])
ax.legend(('14', '20', '26', '32'))
plt.show()
```

```
for df in performance_data:
  df['rolling_val_acc'] = df['val_acc'].rolling(10).sum()
```

```
plt.figure(figsize=(12, 6))
ax=plt.subplot(1,1,1,)
for i in range(len(performance_data)):
  ax.plot(performance_data[i].rolling_val_acc[15:])
ax.legend(('14', '20', '26', '32'))
plt.show()
```
